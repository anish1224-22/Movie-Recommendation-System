```python
import pandas as pd
import numpy as np
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
import ast

def convert(obj):
  if isinstance(obj, str):  # Check if the object is a string
    ans = []
    for i in ast.literal_eval(obj):
      ans.append(i["name"])
    return ans
  else:  # If it's already a list, just return it
    return obj
```

```python
movies['genres']=movies['genres'].apply(convert)
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
  ans=[]
  counter=0
  # Handle NaN values
  if isinstance(obj, float) and math.isnan(obj):
    return ans  # Return an empty list if NaN

  # Check if the object is already a list
  if isinstance(obj, list):
    for i in obj:
      if counter!=3:
        ans.append(i)  # No need to extract 'name' if it's already a string
        counter+=1
      else:
        break
  # If it's not a list, assume it's a string and try to parse it
  else:
    for i in ast.literal_eval(obj):
      if counter!=3:
        ans.append(i['name'])
        counter+=1
      else:
        break
  return ans

import math  # Import math module for isnan check
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetchDirector(obj):
  ans=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director' or i['job']=='Producer':
      ans.append(i['name'])
  return ans
```

```python
movies['crew']=movies['crew'].apply(fetchDirector)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
 def removeSpaces(obj):
   ans=[]
   for i in obj:
     ans.append(i.replace(" ",""))
   return ans
```

```python
movies['genres']=movies['genres'].apply(removeSpaces)
movies['keywords']=movies['keywords'].apply(removeSpaces)
movies['cast']=movies['cast'].apply(removeSpaces)
movies['crew']=movies['crew'].apply(removeSpaces)
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
!pip install scikit-learn
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
!pip install --user pandas tqdm nltk
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
def recommend(movie):
  index=new_df[new_df['title']==movie].index[0]
  distances = similarity[index]
  movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
